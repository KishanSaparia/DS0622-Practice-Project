```python
#lets find the best method for skewness for negative and positive both values
from scipy.stats import boxcox,yeojohnson
from scipy.stats import skew
def skee(a):
    model=[np.sqrt(a),np.log(a),yeojohnson(a)[0]]
    print('original skewness is:',a.skew())
    print('\n')
    for m in model:
        x=m
        print(skew(m))
        print('\n')
```

We have to create to different functions because boxcox is only use in positive values

```python
#lets find the best method for skewness for only positive value
from scipy.stats import boxcox,yeojohnson
from scipy.stats import skew
def skeep(a):
    model=[np.sqrt(a),np.log(a),boxcox(a)[0],yeojohnson(a)[0]]
    print('original skewness is:',a.skew())
    print('\n')
    for m in model:
        x=m
        print(skew(m))
        print('\n')
```

# Since we observe that our target variable is in binary type so we should use classifier model.

now import required basic libraries.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score,roc_curve,accuracy_score
```

For find best random state we can use this for loop.

here we apply stratify because our model has only two values so it is important to split is accordingly.same 0 and 1 has same data

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.2,stratify=y)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print('At random state',i,'The model perform very well')
        print('Random State = ',i)
        print("Training accuracy score is = ",accuracy_score(y_train,pred_train))
        print("Test accuracy score is = ",accuracy_score(y_test,pred_test))
        print('\n')
```
